```python
import os
import math
import zipfile
import random
import urllib.request
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_model_optimization as tfmot

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from tensorflow import keras
from tensorflow.keras import layers

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

print("Python executable:", os.sys.executable)
print("TensorFlow version:", tf.__version__)
print("TF-MOT version:", tfmot.__version__)
```

## 2. Download and Extract the UCI HAR Dataset

The original dataset contains:
- **561 numerical features** extracted from smartphone sensor signals,
- **6 activity classes**, and
- predefined **training** and **test** splits.

The code below downloads and extracts the dataset if it is not already present in the working directory.


```python
dataset_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00240/UCI%20HAR%20Dataset.zip"
zip_path = "uci_har_dataset.zip"
extract_dir = "."

if not os.path.exists("UCI HAR Dataset"):
    urllib.request.urlretrieve(dataset_url, zip_path)
    with zipfile.ZipFile(zip_path, "r") as zf:
        zf.extractall(extract_dir)
    print("Dataset downloaded and extracted.")
else:
    print("Dataset directory already exists.")
```

## 3. Load the Data

```python
def load_har_data(root_dir="UCI HAR Dataset"):
    # 1. Load X_train from train/X_train.txt as float32
    X_train = pd.read_csv(
        os.path.join(root_dir, "train", "X_train.txt"),
        sep=r"\s+", header=None
    ).values.astype("float32")

    # 2. Load y_train from train/y_train.txt as int32
    y_train = pd.read_csv(
        os.path.join(root_dir, "train", "y_train.txt"),
        sep=r"\s+", header=None
    ).values.ravel().astype("int32")

    # 3. Load X_test from test/X_test.txt as float32
    X_test = pd.read_csv(
        os.path.join(root_dir, "test", "X_test.txt"),
        sep=r"\s+", header=None
    ).values.astype("float32")

    # 4. Load y_test from test/y_test.txt as int32
    y_test = pd.read_csv(
        os.path.join(root_dir, "test", "y_test.txt"),
        sep=r"\s+", header=None
    ).values.ravel().astype("int32")

    # 5. Convert labels to zero-based class indices
    y_train = y_train - 1
    y_test = y_test - 1

    return X_train, y_train, X_test, y_test


X_train, y_train, X_test, y_test = load_har_data()

class_names = [
    "WALKING",
    "WALKING_UPSTAIRS",
    "WALKING_DOWNSTAIRS",
    "SITTING",
    "STANDING",
    "LAYING"
]

num_features = X_train.shape[1]
num_classes = len(np.unique(y_train))

print("Training set:", X_train.shape, y_train.shape)
print("Test set    :", X_test.shape, y_test.shape)
print("Input dtype :", X_train.dtype)
print("Label dtype :", y_train.dtype)
print("num_features:", num_features)
print("num_classes :", num_classes)
```

## 4. Quick Inspection

```python
summary_rows = []
for class_idx, class_name in enumerate(class_names):
    summary_rows.append({
        "class_index": class_idx,
        "class_name": class_name,
        "train_samples": int(np.sum(y_train == class_idx)),
    })

class_summary = pd.DataFrame(summary_rows)
print(class_summary.to_string(index=False))
```

## 5. Train a Baseline DNN

We will use a compact dense neural network that is appropriate for a numerical-feature TinyML workflow.

### Architecture
- Input: 561 features
- Dense(256, ReLU)
- Dense(128, ReLU)
- Dense(64, ReLU)
- Dense(6, Softmax)


```python
def build_baseline_model(input_dim, num_classes):
    model = keras.Sequential([
        layers.Input(shape=(input_dim,)),
        layers.Dense(256, activation="relu"),
        layers.Dense(128, activation="relu"),
        layers.Dense(64, activation="relu"),
        layers.Dense(num_classes, activation="softmax")
    ])
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model


baseline_model = build_baseline_model(num_features, num_classes)
baseline_model.summary()
```

### Train the Baseline Model

```python
callbacks = [
    keras.callbacks.EarlyStopping(
        monitor="val_accuracy",
        patience=5,
        restore_best_weights=True
    )
]

history_baseline = baseline_model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=40,
    batch_size=64,
    callbacks=callbacks,
    verbose=1,
)
```

### Training Curves

```python
history_df = pd.DataFrame(history_baseline.history)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

history_df[["accuracy", "val_accuracy"]].plot(ax=axes[0], title="Baseline Training Accuracy")
axes[0].set_xlabel("Epoch")
axes[0].set_ylabel("Accuracy")
axes[0].grid(True, alpha=0.3)

history_df[["loss", "val_loss"]].plot(ax=axes[1], title="Baseline Training Loss")
axes[1].set_xlabel("Epoch")
axes[1].set_ylabel("Loss")
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 6. Evaluate the Baseline Keras Model

```python
# 1. Predicted probabilities on X_test
y_prob_baseline = baseline_model.predict(X_test, verbose=0)

# 2. Convert to class predictions
y_pred_baseline = np.argmax(y_prob_baseline, axis=1)

# 3. Test accuracy
baseline_keras_accuracy = accuracy_score(y_test, y_pred_baseline)
print(f"Baseline Keras test accuracy: {baseline_keras_accuracy:.4f}")

# 4. Classification report
print(classification_report(y_test, y_pred_baseline, target_names=class_names, digits=4))

# 5. Confusion matrix
cm_baseline = confusion_matrix(y_test, y_pred_baseline)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_baseline, display_labels=class_names)

fig, ax = plt.subplots(figsize=(7, 7))
disp.plot(ax=ax, cmap="Blues", colorbar=False, xticks_rotation=45)
ax.set_title("Confusion Matrix - Baseline Keras (FP32)")
plt.tight_layout()
plt.show()
```

# Part I: Model Pruning with Sparsity

In this part, we apply **magnitude-based pruning** to the DNN. The key idea is to gradually set small-magnitude weights to zero during training.

We will compare:
1. the baseline TensorFlow Lite model,
2. the pruned model converted **without** stripping the pruning wrappers, and
3. the stripped sparse model converted with **experimental sparsity-aware optimization**.


## 7. TensorFlow Lite Utilities

The following helper functions are used to:
- convert Keras models to TensorFlow Lite,
- evaluate TensorFlow Lite models on the test set, and
- measure model size.


```python
def save_binary_model(model_content, filename):
    with open(filename, "wb") as f:
        f.write(model_content)
    return os.path.getsize(filename) / 1024.0  # KB

def evaluate_tflite_model(tflite_model, X, y_true):
    interpreter = tf.lite.Interpreter(model_content=tflite_model)
    interpreter.allocate_tensors()

    input_details = interpreter.get_input_details()[0]
    output_details = interpreter.get_output_details()[0]

    input_scale, input_zero_point = input_details["quantization"]
    output_scale, output_zero_point = output_details["quantization"]

    y_pred = []

    for i in range(len(X)):
        x = X[i:i+1].astype(np.float32)

        # Quantize the input only when the TFLite input tensor expects int8 or uint8
        input_dtype = input_details["dtype"]
        if input_dtype in (np.int8, np.uint8):
            if input_scale == 0:
                raise ValueError("Quantized model has invalid input quantization scale 0.")
            x = np.round(x / input_scale + input_zero_point)
            info = np.iinfo(input_dtype)
            x = np.clip(x, info.min, info.max).astype(input_dtype)
        else:
            x = x.astype(input_dtype)

        interpreter.set_tensor(input_details["index"], x)
        interpreter.invoke()

        output = interpreter.get_tensor(output_details["index"])

        # Dequantize the output only when the output tensor is int8 or uint8
        output_dtype = output_details["dtype"]
        if output_dtype in (np.int8, np.uint8) and output_scale != 0:
            output = (output.astype(np.float32) - output_zero_point) * output_scale

        pred = int(np.argmax(output, axis=1)[0])
        y_pred.append(pred)

    acc = accuracy_score(y_true, y_pred)
    return acc, np.array(y_pred)


def convert_to_tflite_fp32(model):
    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    return converter.convert()
```

## 8. Convert the Baseline Model to TensorFlow Lite

```python
tflite_baseline = convert_to_tflite_fp32(baseline_model)

size_baseline = save_binary_model(tflite_baseline, "baseline_model.tflite")
acc_baseline_tflite, y_pred_baseline_tflite = evaluate_tflite_model(tflite_baseline, X_test, y_test)

print(f"Baseline TFLite size    : {size_baseline:.2f} KB")
print(f"Baseline TFLite accuracy: {acc_baseline_tflite:.4f}")
print(f"Baseline Keras accuracy : {baseline_keras_accuracy:.4f}")
```

## 9. Apply Magnitude-Based Pruning

We will prune the DNN using a **polynomial decay schedule**:
- start from low sparsity,
- gradually increase sparsity during training, and
- finish with a highly sparse model.

After training, we will compare:
- the pruned model **with** the pruning wrappers still present, and
- the final sparse model after applying `strip_pruning(...)`.


```python
from tensorflow_model_optimization.sparsity.keras import (
    prune_low_magnitude,
    PolynomialDecay,
    UpdatePruningStep,
    strip_pruning
)

pruning_epochs = 12
batch_size = 64

steps_per_epoch = int(np.ceil(len(X_train) * 0.8 / batch_size))

pruning_params = {
    "pruning_schedule": PolynomialDecay(
        initial_sparsity=0.20,
        final_sparsity=0.85,
        begin_step=0,
        end_step=steps_per_epoch * pruning_epochs,
        frequency=1,
    )
}

print("steps_per_epoch:", steps_per_epoch)
print("end_step :", steps_per_epoch * pruning_epochs)

pruned_model = prune_low_magnitude(
    build_baseline_model(num_features, num_classes),
    **pruning_params
)

pruned_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

pruned_model.summary()

pruning_callbacks = [
    UpdatePruningStep(),
    keras.callbacks.EarlyStopping(
        monitor="val_accuracy",
        patience=3,
        restore_best_weights=True
    )
]

history_pruned = pruned_model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=pruning_epochs,
    batch_size=batch_size,
    callbacks=pruning_callbacks,
    verbose=1,
)
```

## 10. Convert the Pruned Model Before and After Stripping the Pruning Wrappers

First, we convert the pruned model **with** the pruning wrappers still attached.

Next, we strip the pruning wrappers and convert the resulting sparse model with:
- `tf.lite.Optimize.EXPERIMENTAL_SPARSITY`

This is the proper way to preserve sparsity in the exported TensorFlow Lite model.


```python
# 1. Convert the pruned model WITH the pruning wrappers still attached
tflite_pruned_wrapped = convert_to_tflite_fp32(pruned_model)

# 2. Save and evaluate
size_pruned_wrapped = save_binary_model(tflite_pruned_wrapped, "pruned_model_wrapped.tflite")
acc_pruned_wrapped, y_pred_pruned_wrapped = evaluate_tflite_model(tflite_pruned_wrapped, X_test, y_test)

print(f"Pruned (wrappers attached) size    : {size_pruned_wrapped:.2f} KB")
print(f"Pruned (wrappers attached) accuracy: {acc_pruned_wrapped:.4f}")

# 3. Strip the pruning wrappers
pruned_model_stripped = strip_pruning(pruned_model)
pruned_model_stripped.summary()

# 4. Convert the stripped model with EXPERIMENTAL_SPARSITY
converter = tf.lite.TFLiteConverter.from_keras_model(pruned_model_stripped)
converter.optimizations = [tf.lite.Optimize.EXPERIMENTAL_SPARSITY]
tflite_pruned_sparse = converter.convert()

# 5. Save and evaluate
size_pruned_sparse = save_binary_model(tflite_pruned_sparse, "pruned_model_sparse.tflite")
acc_pruned_sparse, y_pred_pruned_sparse = evaluate_tflite_model(tflite_pruned_sparse, X_test, y_test)

print(f"Stripped sparse size    : {size_pruned_sparse:.2f} KB")
print(f"Stripped sparse accuracy: {acc_pruned_sparse:.4f}")
```

## 11. Part I Comparison: Accuracy and Model Size

```python
part1_comparison = pd.DataFrame(
    [
        {"Model": "Baseline", "Format": "FP32 TFLite", "Test Accuracy": round(float(acc_baseline_tflite), 4),   "Model Size (KB)": round(float(size_baseline), 2)},
        {"Model": "Pruned (with mask)", "Format": "FP32 TFLite", "Test Accuracy": round(float(acc_pruned_wrapped), 4),    "Model Size (KB)": round(float(size_pruned_wrapped), 2)},
        {"Model": "Pruned (stripped)", "Format": "FP32 TFLite + sparsity",   "Test Accuracy": round(float(acc_pruned_sparse), 4),     "Model Size (KB)": round(float(size_pruned_sparse), 2)},
    ]
)

print(part1_comparison.to_string(index=False))
```

```python
plt.figure(figsize=(8, 4))
plt.bar(part1_comparison["Model"], part1_comparison["Model Size (KB)"])
plt.ylabel("Model size (KB)")
plt.title("Part I: model size comparison")
plt.xticks(rotation=15, ha="right")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 4))
plt.bar(part1_comparison["Model"], part1_comparison["Test Accuracy"])
plt.ylabel("Accuracy")
plt.ylim(
    max(0.0, part1_comparison["Test Accuracy"].min() - 0.02),
    min(1.0, part1_comparison["Test Accuracy"].max() + 0.02),
)
plt.title("Part I: test accuracy comparison")
plt.xticks(rotation=15, ha="right")
plt.tight_layout()
plt.show()
```

### Confusion Matrix for the Stripped Sparse Model

```python
print(classification_report(y_test, y_pred_pruned_sparse, target_names=class_names, digits=4))

cm_pruned_sparse = confusion_matrix(y_test, y_pred_pruned_sparse)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_pruned_sparse, display_labels=class_names)

fig, ax = plt.subplots(figsize=(7, 7))
disp.plot(ax=ax, cmap="Blues", colorbar=False, xticks_rotation=45)
ax.set_title("Confusion Matrix - Pruned (stripped, sparse FP32)")
plt.tight_layout()
plt.show()
```

# Part II: Model Pruning + Float16 Quantization

In this part, we combine **pruning** and **float16 quantization**.

We will compare:
1. the pruned TensorFlow Lite model **with** the pruning wrappers still attached, after float16 quantization, and
2. the stripped sparse TensorFlow Lite model after **both** sparsity-aware optimization and float16 quantization.

This lets us observe whether properly finalizing the pruned model leads to a more compact deployable representation.


```python
# 1. Convert the pruned model WITH mask to float16
converter = tf.lite.TFLiteConverter.from_keras_model(pruned_model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.target_spec.supported_types = [tf.float16]
tflite_pruned_wrapped_fp16 = converter.convert()

# 2. Save and evaluate
size_pruned_wrapped_fp16 = save_binary_model(tflite_pruned_wrapped_fp16, "pruned_model_wrapped_fp16.tflite")
acc_pruned_wrapped_fp16, y_pred_pruned_wrapped_fp16 = evaluate_tflite_model(tflite_pruned_wrapped_fp16, X_test, y_test)

print(f"Pruned (with mask) + FP16 size    : {size_pruned_wrapped_fp16:.2f} KB")
print(f"Pruned (with mask) + FP16 accuracy: {acc_pruned_wrapped_fp16:.4f}")

# 3. Convert the stripped sparse model to float16
converter = tf.lite.TFLiteConverter.from_keras_model(pruned_model_stripped)
converter.optimizations = [tf.lite.Optimize.DEFAULT, tf.lite.Optimize.EXPERIMENTAL_SPARSITY]
converter.target_spec.supported_types = [tf.float16]
tflite_pruned_sparse_fp16 = converter.convert()

# 4. Save and evaluate
size_pruned_sparse_fp16 = save_binary_model(tflite_pruned_sparse_fp16, "pruned_model_sparse_fp16.tflite")
acc_pruned_sparse_fp16, y_pred_pruned_sparse_fp16 = evaluate_tflite_model(tflite_pruned_sparse_fp16, X_test, y_test)

print(f"Stripped sparse + FP16 size    : {size_pruned_sparse_fp16:.2f} KB")
print(f"Stripped sparse + FP16 accuracy: {acc_pruned_sparse_fp16:.4f}")
```

## 12. Part II Comparison: Accuracy and Model Size

```python
part2_comparison = pd.DataFrame(
    [
        {"Model": "Pruned (with mask)", "Format": "FP32",             "Test Accuracy": round(float(acc_pruned_wrapped), 4),      "Model Size (KB)": round(float(size_pruned_wrapped), 2)},
        {"Model": "Pruned (stripped)",  "Format": "FP32 + sparsity",  "Test Accuracy": round(float(acc_pruned_sparse), 4),       "Model Size (KB)": round(float(size_pruned_sparse), 2)},
        {"Model": "Pruned (with mask)", "Format": "Float16",          "Test Accuracy": round(float(acc_pruned_wrapped_fp16), 4), "Model Size (KB)": round(float(size_pruned_wrapped_fp16), 2)},
        {"Model": "Pruned (stripped)",  "Format": "Float16 + sparsity","Test Accuracy": round(float(acc_pruned_sparse_fp16), 4), "Model Size (KB)": round(float(size_pruned_sparse_fp16), 2)},
    ]
)

print(part2_comparison.to_string(index=False))
```

```python
part2_labels = part2_comparison["Model"] + "\n" + part2_comparison["Format"]

plt.figure(figsize=(9, 4))
plt.bar(part2_labels, part2_comparison["Model Size (KB)"])
plt.ylabel("Model size (KB)")
plt.title("Part II: model size comparison")
plt.xticks(rotation=15, ha="right")
plt.tight_layout()
plt.show()

plt.figure(figsize=(9, 4))
plt.bar(part2_labels, part2_comparison["Test Accuracy"])
plt.ylabel("Accuracy")
plt.ylim(
    max(0.0, part2_comparison["Test Accuracy"].min() - 0.02),
    min(1.0, part2_comparison["Test Accuracy"].max() + 0.02),
)
plt.title("Part II: test accuracy comparison")
plt.xticks(rotation=15, ha="right")
plt.tight_layout()
plt.show()
```

### Confusion Matrix for the Stripped Sparse + Float16 Model

```python
print(classification_report(y_test, y_pred_pruned_sparse_fp16, target_names=class_names, digits=4))

cm_pruned_sparse_fp16 = confusion_matrix(y_test, y_pred_pruned_sparse_fp16)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_pruned_sparse_fp16, display_labels=class_names)

fig, ax = plt.subplots(figsize=(7, 7))
disp.plot(ax=ax, cmap="Blues", colorbar=False, xticks_rotation=45)
ax.set_title("Confusion Matrix - Pruned (stripped) + Float16")
plt.tight_layout()
plt.show()

print("FP16 vs FP32 sparse agreement:", np.mean(y_pred_pruned_sparse_fp16 == y_pred_pruned_sparse))
```

## 13. Summary Questions

Write short answers to the following:
1. Did pruning alone reduce the TensorFlow Lite file size when the pruning wrappers were still attached? No — it doubled it. The pruned model with wrappers attached exported to 1454.18 KB versus the baseline's 726.72 KB, a 2.00× increase. Two things cause this. The wrappers carry 184,968 non-trainable mask variables (one per weight), which the converter serializes alongside the weights themselves. And even setting the masks aside, pruning only sets weights to zero — a zero still occupies 4 bytes in FP32, so the pruned weights compress no better than the originals without an encoding that exploits them.
2. Why does `strip_pruning(...)` matter before export? It removes the PruneLowMagnitude wrappers and bakes the mask's zeros directly into the weight matrices. The model summary confirms this: 370,382 total params (184,968 non-trainable) before stripping, 185,414 params with 0 non-trainable after. Stripping alone drops the file from 1454.18 to 726.72 KB by discarding the masks; pairing it with tf.lite.Optimize.EXPERIMENTAL_SPARSITY drops it further to 555.96 KB by encoding the sparse structure instead of storing runs of zeros verbatim. Both steps are needed. The clearest evidence that stripping is not optional: the pruned+float16 model without stripping came out at 732.59 KB — larger than the 726.72 KB baseline. Two compression techniques applied, negative net compression, purely from skipping this step.
3. Which model had the smallest file size in this notebook? The stripped sparse + float16 model at 365.83 KB, a 1.99× reduction from the 726.72 KB baseline. It required both stripping and float16; neither alone got below 555 KB. Worth noting the two savings don't compound cleanly — stripping is worth 2.62× in FP32 but only 2.00× once float16 is applied, since both are removing partly overlapping redundancy.
4. Did float16 quantization noticeably change the test accuracy? No. Accuracy held at 0.9287 across every Part II variant, and the classification report for the stripped sparse + float16 model was identical to the FP32 sparse one to four decimal places on every class. Float16 retains roughly three significant decimal digits, well beyond what these weights require, so the conversion was effectively lossless.

5. If you were deploying this model on a resource-constrained device, which version would you choose and why? Of the models in this notebook, the stripped sparse + float16 at 365.83 KB and 0.9287 — smallest footprint with no accuracy cost, and the only variant that combines both compression steps correctly. But the honest answer spans both labs: PTQ INT8 from the quantization notebook achieved 185.34 KB at 0.9433 accuracy — half the size and higher accuracy — and provides a fully integer inference path, which matters on an MCU with no FPU. Float16 still requires floating-point support. Pruning here is also under-delivering: EarlyStopping halted training at epoch 2 of 12, so the polynomial schedule reached only 47.4% sparsity rather than the configured 85%, and higher sparsity would improve the sparse encoding's payoff. The general lesson is that pruning and quantization are complements, not substitutes — the strongest MCU candidate would be a fully-pruned model then quantized to INT8, rather than either technique alone.
