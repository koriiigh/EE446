## 4. Load the Data


```python
def load_har_data(root_dir="UCI HAR Dataset"):
    # TODO:
    # 1. Load X_train from train/X_train.txt
    X_train = pd.read_csv(
        os.path.join(root_dir, "train", "X_train.txt"),
        sep=r"\s+", header=None
    ).values.astype("float32")
    # 2. Load y_train from train/y_train.txt
    y_train = pd.read_csv(
        os.path.join(root_dir, "train", "y_train.txt"),
        sep=r"\s+", header=None
    ).values.ravel().astype("int32")
    # 3. Load X_test from test/X_test.txt
    X_test = pd.read_csv(
            os.path.join(root_dir, "test", "X_test.txt"),
            sep=r"\s+", header=None
        ).values.astype("float32")
    # 4. Load y_test from test/y_test.txt
    y_test = pd.read_csv(
        os.path.join(root_dir, "test", "y_test.txt"),
        sep=r"\s+", header=None
    ).values.ravel().astype("int32")
    # 5. Convert the labels to zero-based class indices by subtracting 1
    y_train = y_train - 1
    y_test = y_test - 1

    return X_train, y_train, X_test, y_test

X_train, y_train, X_test, y_test = load_har_data()

class_names = [
    "WALKING",
    "WALKING_UPSTAIRS",
    "WALKING_DOWNSTAIRS",
    "SITTING",
    "STANDING",
    "LAYING"
]

# TODO: define num_features and num_classes
# <-- Enter your code here <--#
num_features = X_train.shape[1]
num_classes = len(np.unique(y_train))

print("Training set:", X_train.shape, y_train.shape)
print("Test set    :", X_test.shape, y_test.shape)
print("Input dtype :", X_train.dtype)
print("Label dtype :", y_train.dtype)
print("num_features:", num_features)
print("num_classes :", num_classes)
```

## 5. Quick Inspection


```python
summary_rows = []
for class_idx, class_name in enumerate(class_names):
    summary_rows.append({
        "class_index": class_idx,
        "class_name": class_name,
        "train_samples": int(np.sum(y_train == class_idx)),
    })

class_summary = pd.DataFrame(summary_rows)
print(class_summary.to_string(index=False))
```

## 6. Train a Baseline DNN

We will use a compact dense neural network that is appropriate for a numerical-feature TinyML-style workflow.

### Architecture
- Input: 561 features
- Dense(256, ReLU)
- Dense(128, ReLU)
- Dense(64, ReLU)
- Dense(6, Softmax)


```python
def build_baseline_model(input_dim, num_classes):
    model = keras.Sequential([
        layers.Input(shape=(input_dim,)),
        layers.Dense(256, activation="relu"),
        layers.Dense(128, activation="relu"),
        layers.Dense(64, activation="relu"),
        layers.Dense(num_classes, activation="softmax")
    ])
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model


baseline_model = build_baseline_model(num_features, num_classes)
baseline_model.summary()
```

### Train the Baseline Model


```python
callbacks = [
    keras.callbacks.EarlyStopping(
        monitor="val_accuracy",
        patience=5,
        restore_best_weights=True
    )
]

history_baseline = baseline_model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=40,
    batch_size=64,
    callbacks=callbacks,
    verbose=1,
)
```

### Training Curves


```python
history_df = pd.DataFrame(history_baseline.history)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

history_df[["accuracy", "val_accuracy"]].plot(ax=axes[0], title="Baseline Training Accuracy")
axes[0].set_xlabel("Epoch")
axes[0].set_ylabel("Accuracy")
axes[0].grid(True, alpha=0.3)

history_df[["loss", "val_loss"]].plot(ax=axes[1], title="Baseline Training Loss")
axes[1].set_xlabel("Epoch")
axes[1].set_ylabel("Loss")
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 7. Evaluate the Baseline Keras Model


```python
# 1. Predicted probabilities on X_test
y_prob_baseline = baseline_model.predict(X_test, verbose=0)

# 2. Convert to class predictions
y_pred_baseline = np.argmax(y_prob_baseline, axis=1)

# 3. Test accuracy
baseline_keras_accuracy = accuracy_score(y_test, y_pred_baseline)
print(f"Baseline Keras test accuracy: {baseline_keras_accuracy:.4f}")

# 4. Classification report
print(classification_report(y_test, y_pred_baseline, target_names=class_names, digits=4))

# 5. Confusion matrix
cm_baseline = confusion_matrix(y_test, y_pred_baseline)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_baseline, display_labels=class_names)

fig, ax = plt.subplots(figsize=(7, 7))
disp.plot(ax=ax, cmap="Blues", colorbar=False, xticks_rotation=45)
ax.set_title("Confusion Matrix - Baseline Keras (FP32)")
plt.tight_layout()
plt.show()
```

## 8. TensorFlow Lite Utilities

The following helper functions are used to:
- convert Keras models to TensorFlow Lite,
- evaluate TensorFlow Lite models on the test set, and
- measure model file size.


```python
def save_binary_model(model_content, filename):
    with open(filename, "wb") as f:
        f.write(model_content)
    return os.path.getsize(filename) / 1024.0  # KB

def representative_dataset_gen():
    for sample in X_train[:300]:
        yield [sample[np.newaxis, ...].astype(np.float32)]

def evaluate_tflite_model(tflite_model, X, y_true):
    interpreter = tf.lite.Interpreter(model_content=tflite_model)
    interpreter.allocate_tensors()

    input_details = interpreter.get_input_details()[0]
    output_details = interpreter.get_output_details()[0]

    input_scale, input_zero_point = input_details["quantization"]
    output_scale, output_zero_point = output_details["quantization"]

    y_pred = []

    for i in range(len(X)):
        x = X[i:i+1].astype(np.float32)

        input_dtype = input_details["dtype"]
        if input_dtype in (np.int8, np.uint8):
            if input_scale == 0:
                raise ValueError("Quantized model has invalid input quantization scale 0.")
            x = np.round(x / input_scale + input_zero_point)
            info = np.iinfo(input_dtype)
            x = np.clip(x, info.min, info.max).astype(input_dtype)
        else:
            x = x.astype(input_dtype)

        interpreter.set_tensor(input_details["index"], x)
        interpreter.invoke()

        output = interpreter.get_tensor(output_details["index"])

        output_dtype = output_details["dtype"]
        if output_dtype in (np.int8, np.uint8) and output_scale != 0:
            output = (output.astype(np.float32) - output_zero_point) * output_scale

        y_pred.append(np.argmax(output, axis=1)[0])

    y_pred = np.array(y_pred)
    acc = accuracy_score(y_true, y_pred)
    return acc, y_pred

def convert_to_tflite_fp32(model):
    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    return converter.convert()

def convert_to_tflite_dynamic_range(model):
    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    converter.optimizations = [tf.lite.Optimize.DEFAULT]
    return converter.convert()

def convert_to_tflite_float16(model):
    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    converter.optimizations = [tf.lite.Optimize.DEFAULT]
    converter.target_spec.supported_types = [tf.float16]
    return converter.convert()


def convert_to_tflite_int8(model):
    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    converter.optimizations = [tf.lite.Optimize.DEFAULT]
    converter.representative_dataset = representative_dataset_gen
    converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
    converter.inference_input_type = tf.int8
    converter.inference_output_type = tf.int8
    return converter.convert()
```

## 9. Post-Training Quantization (PTQ)


```python
ptq_models = [
    ("Baseline TFLite",   "model_fp32.tflite", convert_to_tflite_fp32(baseline_model)),
    ("PTQ INT8", "model_int8.tflite", convert_to_tflite_int8(baseline_model)),
    ("PTQ Dynamic Range", "model_dynamic_range.tflite", convert_to_tflite_dynamic_range(baseline_model)),
    ("PTQ Float16", "model_float16.tflite", convert_to_tflite_float16(baseline_model)),
]

ptq_rows = []
ptq_predictions = {}

for model_name, filename, tflite_model in ptq_models:
    size_kb = save_binary_model(tflite_model, filename)
    acc, y_pred = evaluate_tflite_model(tflite_model, X_test, y_test)
    ptq_predictions[model_name] = y_pred

    ptq_rows.append(
        {
            "model": model_name,
            "file": filename,
            "size_kb": size_kb,
            "accuracy": acc,
            "accuracy_drop_vs_keras_baseline": baseline_keras_accuracy - acc,
        }
    )

ptq_results = pd.DataFrame(ptq_rows)


keras_reference = pd.DataFrame(
    [
        {
            "model": "Baseline Keras",
            "file": "-",
            "size_kb": np.nan,
            "accuracy": round(float(baseline_keras_accuracy), 4),
            "accuracy_drop_vs_keras_baseline": 0.0,
        }
    ]
)

ptq_results = pd.concat([keras_reference, ptq_results], ignore_index=True)
ptq_results["size_kb"] = ptq_results["size_kb"].round(2)
ptq_results["accuracy"] = ptq_results["accuracy"].round(4)
ptq_results["accuracy_drop_vs_keras_baseline"] = ptq_results["accuracy_drop_vs_keras_baseline"].round(4)

print(ptq_results.to_string(index=False))
```

## 10. PTQ Comparison: Accuracy and Model Size


```python
ptq_comparison = pd.DataFrame(
    [
        {"Model Family": "Keras",  "Format": "FP32 (Keras)", "Test Accuracy": baseline_keras_accuracy, "Model Size (KB)": np.nan},
        {"Model Family": "TFLite", "Format": "FP32", "Test Accuracy": ptq_results.loc[ptq_results["model"] == "Baseline TFLite",   "accuracy"].item(), "Model Size (KB)": ptq_results.loc[ptq_results["model"] == "Baseline TFLite",   "size_kb"].item()},
        {"Model Family": "PTQ", "Format": "Dynamic Range", "Test Accuracy": ptq_results.loc[ptq_results["model"] == "PTQ Dynamic Range", "accuracy"].item(), "Model Size (KB)": ptq_results.loc[ptq_results["model"] == "PTQ Dynamic Range", "size_kb"].item()},
        {"Model Family": "PTQ", "Format": "Float16", "Test Accuracy": ptq_results.loc[ptq_results["model"] == "PTQ Float16", "accuracy"].item(), "Model Size (KB)": ptq_results.loc[ptq_results["model"] == "PTQ Float16", "size_kb"].item()},
        {"Model Family": "PTQ", "Format": "INT8", "Test Accuracy": ptq_results.loc[ptq_results["model"] == "PTQ INT8", "accuracy"].item(), "Model Size (KB)": ptq_results.loc[ptq_results["model"] == "PTQ INT8", "size_kb"].item()},
    ]
)

print(ptq_comparison.to_string(index=False))
```

```python
# Visual comparison of PTQ model sizes
ptq_plot_df = ptq_comparison.dropna(subset=["Model Size (KB)"]).copy()

plt.figure(figsize=(8, 4))
plt.bar(ptq_plot_df["Format"], ptq_plot_df["Model Size (KB)"])
plt.ylabel("Model size (KB)")
plt.title("PTQ model size comparison")
plt.xticks(rotation=15, ha="right")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 4))
plt.bar(ptq_comparison["Format"], ptq_comparison["Test Accuracy"])
plt.ylabel("Accuracy")
plt.ylim(
    max(0.0, ptq_comparison["Test Accuracy"].min() - 0.02),
    min(1.0, ptq_comparison["Test Accuracy"].max() + 0.02),
)
plt.title("PTQ accuracy comparison")
plt.xticks(rotation=15, ha="right")
plt.tight_layout()
plt.show()
```

### Confusion Matrix for the PTQ Int8 Model


```python
y_pred_int8 = ptq_predictions["PTQ INT8"]

print(classification_report(y_test, y_pred_int8, target_names=class_names, digits=4))

cm_int8 = confusion_matrix(y_test, y_pred_int8)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_int8, display_labels=class_names)

fig, ax = plt.subplots(figsize=(7, 7))
disp.plot(ax=ax, cmap="Blues", colorbar=False, xticks_rotation=45)
ax.set_title("Confusion Matrix - PTQ INT8")
plt.tight_layout()
plt.show()

print("INT8 vs Keras baseline prediction agreement:", np.mean(y_pred_int8 == y_pred_baseline))
```

## 11. Quantization-Aware Training (QAT)

QAT simulates quantization effects during training so that the final model is usually more robust after conversion to int8.


```python
qat_model = tfmot.quantization.keras.quantize_model(baseline_model)

qat_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-4),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
)

qat_model.summary()
```

### Fine-Tune the QAT Model


```python
history_qat = qat_model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=8,
    batch_size=64,
    verbose=1,
)
```

### Evaluate the QAT Keras Model


```python
qat_loss, qat_accuracy = qat_model.evaluate(X_test, y_test, verbose=0)
print(f"QAT Keras test accuracy: {qat_accuracy:.4f}")

y_prob_qat = qat_model.predict(X_test, verbose=0)
y_pred_qat = np.argmax(y_prob_qat, axis=1)

print(classification_report(y_test, y_pred_qat, target_names=class_names, digits=4))
```

### Convert the QAT Model to Int8 TensorFlow Lite


```python
tflite_qat_int8 = convert_to_tflite_int8(qat_model)

size_qat_int8 = save_binary_model(tflite_qat_int8, "model_qat_int8.tflite")
acc_qat_int8, y_pred_qat_int8 = evaluate_tflite_model(tflite_qat_int8, X_test, y_test)

print(f"QAT INT8 TFLite size    : {size_qat_int8:.2f} KB")
print(f"QAT INT8 TFLite accuracy: {acc_qat_int8:.4f}")
```

## 13. Summary Questions

Write short answers to the following:
1. Which quantization method gave the smallest model size? PTQ INT8, at 185.34 KB — a 3.92× reduction from the 726.72 KB FP32 TFLite baseline. Dynamic range was effectively tied at 186.11 KB, since both store weights as 8-bit integers; the 0.77 KB difference is metadata. Float16 was only 2× smaller (365.71 KB), as expected from halving 32-bit weights to 16 bits.
2. Which quantization method gave the best accuracy among the TensorFlow Lite models? Dynamic range, at 0.9440 vs 0.9433 for FP32, Float16, and PTQ INT8. But this difference is not meaningful: 0.0007 × 2,947 ≈ 2 samples. Quantization noise nudged two borderline predictions across a decision boundary in the favorable direction. All four PTQ variants should be treated as accuracy-equivalent; only QAT INT8 (0.9382) is measurably lower.
3. Did QAT improve the final int8 model compared with PTQ int8? No — QAT INT8 scored 0.9382 vs PTQ INT8's 0.9433, at the same file size (185.70 vs 185.34 KB). The cause is not quantization. Comparing each INT8 model to its own FP32 parent: PTQ lost 0.0000 (0.9433 → 0.9433) and QAT lost 0.0004 (0.9386 → 0.9382, a single sample out of 2,947). Both conversions were essentially lossless. QAT's purpose is to close the FP32→INT8 gap, but PTQ's gap was already zero, so there was nothing to close. The 8 epochs of QAT fine-tuning instead pushed the model past the baseline's early-stopped optimum — training accuracy rose to 0.9889 while val_loss drifted from 0.1126 to 0.1270 — costing ~0.5% generalization on a subject-wise test split. QAT INT8 is worse because the model it started from was worse, not because of quantization.
4. Why is this dataset a good fit for a DNN-based TinyML workflow? Each sample is already a fixed-length vector of 561 engineered features, so a plain feedforward network is the natural architecture — no convolution or recurrence needed. The classes are well balanced (986–1,407 training samples each), so accuracy is a fair metric. Most importantly, the features derive from accelerometer and gyroscope signals, exactly the sensors on-board a device like the Nano 33 BLE Sense — so the inference could realistically run where the data is captured. The resulting model is small enough to matter: 185 KB fits comfortably in MCU flash.
5. If you were deploying this model on a resource-constrained device, which version would you choose and why? PTQ INT8. It is the smallest (185.34 KB, 3.92× compression), matches the FP32 baseline exactly (0.9433), and — critically — is the only variant with a fully integer inference path (inference_input_type and inference_output_type both tf.int8), so it runs on an MCU with no floating-point unit. Dynamic range is the same size but keeps activations in float, and Float16 is twice as large for no accuracy gain. QAT INT8 costs accuracy for no size benefit. This follows the cookbook's recommended workflow: try PTQ INT8 first, escalate to QAT only if accuracy loss is too large — here the loss was zero, so PTQ was sufficient.
